```python
# Pandas and Numpy to process and read the tabular data
import pandas as pd
import numpy as np

# OS commands to extract files/ names in a directory
from os import listdir
from os.path import isfile, join, splitext
```

```python
# Reading in the original dataset
df = pd.read_csv("Datasets/Dataset_Original.csv", index_col="ID")

# Inspecting the general structure of the dataset
df.head()
```

```python
# We created a seperate .csv file for the current value of a column and its desired value

# Yield filenames and extension from the specified directory
onlyfiles = [f for f in listdir('Datasets/old-new/') if isfile(join('Datasets/old-new/', f))]

# We will keep the original values for Impulsive and Sensation_seeking because these are usefully preprocessed for our testing and path coefficient estimation task (see EDA)

# First loop will iterate over all the different .csv files in the directory
for file in onlyfiles:
    new_old = pd.read_csv('Datasets/old-new/'+ file) # Read it in as a Pandas dataframe
    new_old_arr = new_old.to_numpy() # Transform the dataframe to a Numpy array
    filename_raw = splitext(file)[0] # The filename without the extension (without the .csv part)
    for old, new in new_old_arr: # Second loop will iterate through the individual arrays/ .csv files
        df[filename_raw] = df[filename_raw].replace([old],new) # Checks which column to update - replaces all the current values to its respective new value
```

```python
# Confirm that the data has been processed
df.head()
```

```python
# Export the processed dataframe for the prediction task - dataset with numerical columns that are normally distributed
df.to_csv(r'Datasets/Dataset_Coeff_Estimation.csv',index=False)

# Using the same technique there is also a dataset created that uses all the original values (so there we will not keep the preprocessed Impulsive and Sensation_seeking variables)
```
